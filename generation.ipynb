```python
import faker
import pandas as pd
from sqlalchemy import create_engine,text
import random
import mysql.connector 
```

```python
# create a connection with a mysql database
engine = create_engine('mysql+mysqlconnector://root:root@localhost/db_restaurante')
```

```python
def insert_empresa():
    fake = faker.Faker('pt_BR')
    data = []
    num_funcionarios = random.randint(1, 10)  # Número aleatório de funcionários entre 1 e 10
    empresa = fake.company()
    for _ in range(num_funcionarios):  # Adicionar funcionários para a empresa
            data.append({
                'nome_empresa': empresa,
                'nome_funcionario_empresa': fake.name(),
            })
    df = pd.DataFrame(data)
    df.to_sql('st_empresa', con=engine, if_exists='append', index=False)
    return data
```

```python
insert_empresa()
```

```python
def insert_clients():
    fake = faker.Faker('pt_BR')
    funcionarios = insert_empresa()
    
    # Extrair nomes dos funcionários
    nomes_funcionarios = [f['nome_funcionario_empresa'] for f in funcionarios]
    
    # Gerar clientes adicionais (sem repetir os nomes dos funcionários)
    num_clientes_total = 50
    num_clientes_adicionais = num_clientes_total - len(nomes_funcionarios)
    
    data = []
    
    # Adicionar funcionários como clientes
    for nome_funcionario in nomes_funcionarios:
        data.append({
            'cpf_cliente': fake.cpf(),
            'nome_cliente': nome_funcionario,
            'telefone_cliente': fake.phone_number(),
            'email_cliente': fake.email()
        })
    
    # Adicionar clientes aleatórios
    for _ in range(num_clientes_adicionais):
        data.append({
            'cpf_cliente': fake.cpf(),
            'nome_cliente': fake.name(),
            'telefone_cliente': fake.phone_number(),
            'email_cliente': fake.email()
        })
        
    df = pd.DataFrame(data)
    df.to_sql('st_cliente', con=engine, if_exists='append', index=False)
    return data
```

```python
insert_clients()
```

```python
clientes_gerados = insert_clients()

# Exibir os nomes dos clientes gerados
print("Nomes dos clientes gerados:")
for cliente in clientes_gerados:
    print(cliente['nome_cliente'])
```
